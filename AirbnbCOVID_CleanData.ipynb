```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Store filepaths in dataframes
Airbnb_df = "data/AirBNB.csv"
Airbnb_df = pd.read_csv(Airbnb_df)
Airbnb_df.head()
```

```python
len(Airbnb_df)
```

```python
#Drop duplicate scrapes from 1/4/20, 11/8/17, 12/7/17 & 1/17/18
Airbnb_clean_df_r01_04_2020 = Airbnb_df[Airbnb_df.last_scraped != "1/4/2020"]
Airbnb_clean_df_r11_08_2017 = Airbnb_clean_df_r01_04_2020[Airbnb_clean_df_r01_04_2020.last_scraped != "11/8/2017"]
Airbnb_clean_df_r12_07_2017 = Airbnb_clean_df_r11_08_2017[Airbnb_clean_df_r11_08_2017.last_scraped != "12/7/2017"]
Airbnb_clean_df = Airbnb_clean_df_r12_07_2017[Airbnb_clean_df_r12_07_2017.last_scraped != "1/17/2018"]

#Pull just the columns you need
Airbnb_clean_df = Airbnb_clean_df[['id','last_scraped','neighbourhood','price','room_type','zipcode','latitude','longitude']]

len(Airbnb_clean_df)
```

```python
#Convert last_scaped to date format
Airbnb_clean_df['last_scraped'] = pd.to_datetime(Airbnb_clean_df['last_scraped'], errors='coerce')

#Split the data by month & year and place into new columns
Airbnb_clean_df['month'] = Airbnb_clean_df['last_scraped'].dt.month
Airbnb_clean_df['year'] = Airbnb_clean_df['last_scraped'].dt.year

# Drop years you do not need (anything prior to 2017)
Airbnb_clean_df_2015 = Airbnb_clean_df[Airbnb_clean_df.year != 2015]
Airbnb_clean_df = Airbnb_clean_df_2015[Airbnb_clean_df_2015.year != 2016]
Airbnb_clean_df
```

```python
# Convert month & year integers to strings so you can combine
Airbnb_clean_df['year'] = Airbnb_clean_df['year'].astype(str)
Airbnb_clean_df['month'] = Airbnb_clean_df['month'].astype(str)

# Combine month & year in one column and add a zero
Airbnb_clean_df['month_scraped'] = Airbnb_clean_df['year'].str.cat(Airbnb_clean_df['month'],sep="-0")

# Now remove the zero from double digit months
Airbnb_clean_df['month_scraped'] = Airbnb_clean_df['month_scraped'].replace({
    "2020-012":"2020-12","2019-012":"2019-12","2018-012":"2018-12","2017-012":"2017-12",
    "2020-011":"2020-11","2019-011":"2019-11","2018-011":"2018-11","2017-011":"2017-11",
    "2020-010":"2020-10","2019-010":"2019-10","2018-010":"2018-10","2017-010":"2017-10"
})
Airbnb_clean_df
```

```python
# Sort by price to identify any abnormalities
Airbnb_clean_df_sorted = Airbnb_clean_df.sort_values(by=['price'], ascending=False)
Airbnb_clean_df_sorted[:20]
```

```python
# Drop junk data listing
Airbnb_clean_df = Airbnb_clean_df[Airbnb_clean_df.id != 18448377]
len(Airbnb_clean_df)
```
